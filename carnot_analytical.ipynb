$$
\begin{equation}
    \begin{split}
         \frac{dT_m}{dt} & = \frac{1}{c_5} [F'\tau \alpha K_{dir} I_{dir} - c_1 (T_m - T_{amb}) - c_2 (T_m - T_{amb})^2 -2 \frac{\dot m C_p}{A} (T_{m} - T_{in})]
    \end{split}
\end{equation}
$$

```python
from sympy import *
init_printing()

T_m = Function('T_m')
t = symbols('t', real=True, positive=True)
c1 = symbols('c_1', real=True, positive=True)
c2 = symbols('c_2', real=True, positive=True)
c5 = symbols('c_5', real=True, positive=True)
cp = symbols('C_p', real=True, positive=True)
mdot = symbols('mdot', real=True, positive=True)
fta = symbols(r"F'\tau\alpha", real=True, positive=True)

T_amb = symbols('T_{amb}', real=True, positive=True)
T_out = symbols('T_{out}', real=True, positive=True)
T_in = symbols('T_{in}', real=True, positive=True)
A = symbols('A', real=True, positive=True)
k_dir = symbols('K_{dir}', real=True, positive=True)
I_dir = symbols('I_{dir}', real=True, positive=True)

eq = Eq(T_m(t).diff(t), (fta*k_dir*I_dir -c1*(T_m(t) -T_amb) -c2*(T_m(t) -T_amb)**2 -2*mdot*cp*(T_m(t) - T_in)/A)/c5)
eq
```

```python
sol = dsolve(eq, T_m(t)).simplify()
sol
```
